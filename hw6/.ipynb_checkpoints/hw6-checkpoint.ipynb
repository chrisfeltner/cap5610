```python
import pandas as pd
from surprise import Reader
from surprise import Dataset
from surprise.prediction_algorithms.matrix_factorization import SVD
from surprise import KNNBasic
from surprise.model_selection import cross_validate
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../data/movies/ratings_small.csv")
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df[['userId', 'movieId', 'rating']], reader)
```

# Comparing PMF, User-Based CF, and Item-Based CF

```python
algo = SVD(biased=False) #PMF
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
sim_options = {
    'user_based': True
}
algo = KNNBasic(sim_options = sim_options)
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
sim_options = {
    'user_based': False # item based
}
algo = KNNBasic(sim_options = sim_options)
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

# Comparing Metrics: User-Based

```python
mae_means = []
rmse_means = []
```

```python
sim_options = {
    'user_based': True,
    'name': 'cosine'
}
algo = KNNBasic(sim_options = sim_options)
results = cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
mae_means.append(results['test_mae'].mean())
rmse_means.append(results['test_rmse'].mean())
```

```python
sim_options = {
    'user_based': True,
    'name': 'MSD'
}
algo = KNNBasic(sim_options = sim_options)
results = cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
mae_means.append(results['test_mae'].mean())
rmse_means.append(results['test_rmse'].mean())
```

```python
sim_options = {
    'user_based': True,
    'name': 'pearson'
}
algo = KNNBasic(sim_options = sim_options)
results = cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
mae_means.append(results['test_mae'].mean())
rmse_means.append(results['test_rmse'].mean())
```

```python
measures = ['Cosine', 'MSD', 'Pearson']
plt.bar(measures, mae_means)
plt.xlabel('Similarity Metric')
plt.ylabel('Mean MAE')
plt.title("User-Based CF: MAE vs Similiarity Metric")
plt.show()
```

```python
plt.bar(measures, rmse_means)
plt.xlabel('Similarity Metric')
plt.ylabel('Mean RMSE')
plt.title("User-Based CF: RMSE vs Similiarity Metric")
plt.show()
```

# Finding Best K: User-Based

```python
sim_options = {
    'user_based': True,
    'name': 'MSD'
}
```

```python
mae_means = []
rmse_means = []
ks = []
for k in range(2, 101):
    print('============================='+str(k)+" neighbors ========================")
    algo = KNNBasic(sim_options = sim_options, k = k)
    results = cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=False)
    mae_means.append(results['test_mae'].mean())
    rmse_means.append(results['test_rmse'].mean())
    ks.append(k)
```

```python
plt.plot(ks, mae_means)
plt.xlabel('K')
plt.ylabel('Mean MAE')
plt.title("User-Based CF: MAE vs K")
plt.show()
```

```python
plt.plot(ks, rmse_means)
plt.xlabel('K')
plt.ylabel('Mean RMSE')
plt.title("User-Based CF: RMSE vs K")
plt.show()
```

```python
print("Best K: User-Based")
print(ks[np.argmin(rmse_means)])
```
